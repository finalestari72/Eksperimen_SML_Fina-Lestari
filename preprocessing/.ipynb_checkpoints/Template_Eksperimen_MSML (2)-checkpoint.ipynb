# **1. Perkenalan Dataset**


Tahap pertama, Anda harus mencari dan menggunakan dataset dengan ketentuan sebagai berikut:

### Deskripsi Dataset

- **Nama Dataset:** Water Potability
- **Sumber:** Kaggle
- **Link:** https://www.kaggle.com/datasets/adityakadiwal/water-potability
- **Jumlah Baris:** 3.276
- **Jumlah Kolom:** 10
- **Target/Label:** `Potability` (0 = tidak layak minum, 1 = layak minum)
- **Deskripsi Singkat:**  
  Dataset ini berisi informasi kualitas air dari berbagai sumber. Setiap baris merepresentasikan sampel air dengan 9 parameter kimia/fisika. Tujuannya adalah memprediksi apakah air tersebut **layak diminum** (potable) berdasarkan parameter-parameter tersebut.

### Penjelasan Fitur

| Fitur | Deskripsi | Satuan |
|---|---|---|
| `ph` | Tingkat keasaman air | 0–14 |
| `Hardness` | Kapasitas air mengendapkan sabun | mg/L |
| `Solids` | Total dissolved solids | ppm |
| `Chloramines` | Jumlah kloramin | ppm |
| `Sulfate` | Jumlah sulfat terlarut | mg/L |
| `Conductivity` | Konduktivitas listrik air | μS/cm |
| `Organic_carbon` | Jumlah karbon organik | ppm |
| `Trihalomethanes` | Jumlah trihalometan | μg/L |
| `Turbidity` | Ukuran kekeruhan air | NTU |
| `Potability` | **Target** — layak minum (1) / tidak (0) | — |

# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

Tujuan dari EDA adalah untuk memperoleh wawasan awal yang mendalam mengenai data dan menentukan langkah selanjutnya dalam analisis atau pemodelan.

```python
# 4.1 Statistik Deskriptif 
print('=' * 55)
print('STATISTIK DESKRIPTIF')
print('=' * 55)
df.describe().T.style.background_gradient(cmap='Blues', subset=['mean', 'std'])
```

```python
# 4.2 Distribusi Target (Potability)
target_counts = df['Potability'].value_counts()
labels = ['Tidak Layak Minum (0)', 'Layak Minum (1)']
colors = ['#E07B7B', '#6BAF92']

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# Bar chart
axes[0].bar(labels, target_counts.values, color=colors, edgecolor='white', width=0.5)
axes[0].set_title('Distribusi Kelas Target', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Jumlah Sampel')
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 30, str(v), ha='center', fontweight='bold')

# Pie chart
axes[1].pie(target_counts.values, labels=labels, colors=colors,
            autopct='%1.1f%%', startangle=140,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Proporsi Kelas Target', fontsize=13, fontweight='bold')

plt.suptitle('Analisis Class Balance — Potability', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()

print(f'\nRasio kelas: {target_counts[0]} tidak layak : {target_counts[1]} layak')
print(f'Class imbalance ratio: {target_counts[0]/target_counts[1]:.2f}:1')
```

```python
# 4.3 Analisis Missing Values
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': missing_pct
}).sort_values('Persentase (%)', ascending=False)

print('=' * 45)
print('MISSING VALUES PER KOLOM')
print('=' * 45)
print(missing_df)

# Visualisasi
fig, ax = plt.subplots(figsize=(10, 4))
bars = ax.barh(
    missing_df[missing_df['Jumlah Missing'] > 0].index,
    missing_df[missing_df['Jumlah Missing'] > 0]['Persentase (%)'],
    color='#E07B7B', edgecolor='white'
)
ax.set_xlabel('Missing (%)')
ax.set_title('Persentase Missing Values per Kolom', fontsize=13, fontweight='bold')
for bar in bars:
    w = bar.get_width()
    ax.text(w + 0.2, bar.get_y() + bar.get_height()/2,
            f'{w:.1f}%', va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 4.4 Distribusi Setiap Fitur Numerik
features = [c for c in df.columns if c != 'Potability']

fig, axes = plt.subplots(3, 3, figsize=(15, 11))
axes = axes.flatten()

for i, col in enumerate(features):
    data = df[col].dropna()
    axes[i].hist(data, bins=40, color='steelblue', edgecolor='white', alpha=0.85)
    axes[i].axvline(data.mean(), color='red', linestyle='--', linewidth=1.5, label=f'Mean: {data.mean():.2f}')
    axes[i].axvline(data.median(), color='orange', linestyle='--', linewidth=1.5, label=f'Median: {data.median():.2f}')
    axes[i].set_title(col, fontsize=11, fontweight='bold')
    axes[i].legend(fontsize=8)
    axes[i].set_ylabel('Frekuensi')

plt.suptitle('Distribusi Fitur — Water Potability', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
# 4.5 Boxplot — Deteksi Outlier per Kelas
fig, axes = plt.subplots(3, 3, figsize=(15, 11))
axes = axes.flatten()

for i, col in enumerate(features):
    groups = [df[df['Potability'] == 0][col].dropna(),
              df[df['Potability'] == 1][col].dropna()]
    bp = axes[i].boxplot(groups, patch_artist=True, widths=0.5,
                          boxprops=dict(linewidth=1.5),
                          medianprops=dict(color='black', linewidth=2))
    bp['boxes'][0].set_facecolor('#E07B7B')
    bp['boxes'][1].set_facecolor('#6BAF92')
    axes[i].set_xticks([1, 2])
    axes[i].set_xticklabels(['Tidak Layak (0)', 'Layak (1)'], fontsize=9)
    axes[i].set_title(col, fontsize=11, fontweight='bold')

plt.suptitle('Boxplot Fitur per Kelas Potability', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
# 4.6 Correlation Heatmap
plt.figure(figsize=(11, 8))
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr, mask=mask, annot=True, fmt='.2f',
    cmap='RdYlGn', center=0, square=True,
    linewidths=0.5, cbar_kws={'shrink': 0.8},
    annot_kws={'size': 9}
)
plt.title('Correlation Heatmap — Water Quality Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Korelasi dengan target
print('\nKorelasi Fitur terhadap Target (Potability):')
print(corr['Potability'].drop('Potability').sort_values(ascending=False))
```

```python
# 4.7 Pairplot (3 fitur paling berkorelasi dengan target)
top_features = corr['Potability'].drop('Potability').abs().nlargest(3).index.tolist()
top_features.append('Potability')

sns.pairplot(
    df[top_features].dropna(),
    hue='Potability',
    palette={0: '#E07B7B', 1: '#6BAF92'},
    plot_kws={'alpha': 0.5, 's': 20},
    diag_kind='kde'
)
plt.suptitle('Pairplot — Top 3 Fitur vs Potability', y=1.02, fontsize=13)
plt.show()
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Berikut adalah tahapan-tahapan yang bisa dilakukan, tetapi **tidak terbatas** pada:
1. Menghapus atau Menangani Data Kosong (Missing Values)
2. Menghapus Data Duplikat
3. Normalisasi atau Standarisasi Fitur
4. Deteksi dan Penanganan Outlier
5. Encoding Data Kategorikal
6. Binning (Pengelompokan Data)

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
# Buat salinan agar data asli tidak berubah
df_clean = df.copy()
print(f'Salinan dataset dibuat. Shape awal: {df_clean.shape}')
```

```python
# 5.1 Hapus Duplikat
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
print(f'Duplikat dihapus : {before - len(df_clean)} baris')
print(f'Shape            : {df_clean.shape}')
```

```python
# 5.2 Imputasi Missing Values dengan Median
# Median dipilih karena distribusi beberapa fitur sedikit skewed
cols_with_missing = ['ph', 'Sulfate', 'Trihalomethanes']

imputer = SimpleImputer(strategy='median')
df_clean[cols_with_missing] = imputer.fit_transform(df_clean[cols_with_missing])

print(f'Imputasi median pada: {cols_with_missing}')
print(f'Missing values tersisa: {df_clean.isnull().sum().sum()}')

# Visualisasi distribusi sebelum vs sesudah imputasi
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, col in enumerate(cols_with_missing):
    axes[i].hist(df[col].dropna(), bins=30, alpha=0.6, label='Sebelum', color='#E07B7B')
    axes[i].hist(df_clean[col], bins=30, alpha=0.6, label='Sesudah', color='#6BAF92')
    axes[i].set_title(f'Imputasi: {col}', fontweight='bold')
    axes[i].legend()
plt.suptitle('Distribusi Sebelum vs Sesudah Imputasi Missing Values', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# 5.3 Penanganan Outlier (IQR Method)
features = [c for c in df_clean.columns if c != 'Potability']
before   = len(df_clean)

for col in features:
    Q1  = df_clean[col].quantile(0.25)
    Q3  = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    removed = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    if removed > 0:
        print(f'  [{col}] Outlier: {removed} baris  (batas: {lower:.2f} – {upper:.2f})')
    df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

print(f'\nTotal dihapus : {before - len(df_clean)} baris')
print(f'Shape sesudah : {df_clean.shape}')
```

```python
# 5.4 Normalisasi Fitur (StandardScaler)
# Pisahkan fitur dan target
X = df_clean.drop('Potability', axis=1)
y = df_clean['Potability']

scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

print('Normalisasi StandardScaler diterapkan')
print(f'\nStatistik fitur setelah scaling (Mean ≈ 0, Std ≈ 1):')
print(X_scaled.describe().loc[['mean', 'std']].round(4))
```

```python
# 5.5 Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print('Train-Test Split selesai')
print(f'   Train : {X_train.shape[0]} sampel ({X_train.shape[0]/len(X_scaled)*100:.1f}%)')
print(f'   Test  : {X_test.shape[0]} sampel ({X_test.shape[0]/len(X_scaled)*100:.1f}%)')
print(f'\nDistribusi kelas train : {dict(y_train.value_counts())}')
print(f'Distribusi kelas test  : {dict(y_test.value_counts())}')
```

```python
# 5.6 Ringkasan Preprocessing
print('=' * 55)
print('RINGKASAN PREPROCESSING')
print('=' * 55)
print(f'Shape awal            : {df.shape}')
print(f'Shape akhir (X)       : {X_scaled.shape}')
print(f'Missing values        : {df_clean.isnull().sum().sum()}')
print(f'Duplikat              : {df_clean.duplicated().sum()}')
print(f'Fitur yang digunakan  : {list(X_scaled.columns)}')
print(f'Target                : Potability')
print(f'Metode imputasi       : Median (ph, Sulfate, Trihalomethanes)')
print(f'Outlier handling      : IQR Method (factor=1.5)')
print(f'Normalisasi           : StandardScaler')
print(f'Split ratio           : 80% train / 20% test')
```

```python
# Simpan Dataset yang Sudah Diproses 
OUTPUT_DIR  = '../preprocessing'
OUTPUT_PATH = os.path.join(OUTPUT_DIR, 'water_potability_preprocessing.csv')

# Gabungkan kembali X dan y
df_final = X_scaled.copy()
df_final['Potability'] = y.values

df_final.to_csv(OUTPUT_PATH, index=False)

print(f'Dataset preprocessing disimpan ke: {OUTPUT_PATH}')
print(f'Total data siap latih: {df_final.shape[0]} baris × {df_final.shape[1]} kolom')
df_final.head()
```
